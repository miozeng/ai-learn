### 2.2 计算用户与用户之间的相似度
我们可以通过欧几里得距离计算用户A B C D E 两两之间的距离 

也可以通过皮尔森相关系数计算用户两两直接的相关系数

距离越小或者相关系数越大则证明两者之间的关系越强越相似

#### 2.2.1 皮尔森计算方式

```python
def UserSimilarity(user_item):
    item_users = {}  #user 与user之间的相关系数
    C = {} #User 和User的相似矩阵
    for indexs in user_item_pd.columns:
        C.setdefault(indexs, {})
        for indexsc in user_item_pd.columns:
            if indexs ==indexsc:
                continue
            pccs = np.corrcoef(user_item_pd[indexs], user_item_pd[indexsc])
            #print(pccs[0][1])
            C[indexs].setdefault(indexsc,0)
            C[indexs][indexsc] =np.abs(pccs[0][1])
    return C 
```

#### 2.2.2 欧式距离
np.sqrt(np.sum(np.square(x-y)))

```python
def UserSimilarity2(user_item):
    item_users = {}  #user 与user之间的相关系数
    C = {} #User 和User的相似矩阵
    for user, items in user_item.items():
        C.setdefault(user, {})
        for uc, ic in user_item.items():
            #自己不需要计算自己的
            sum = 0 
            if uc == user:
                continue
            #计算欧氏距离
            for i in items.keys():
                if i in ic:
                    sum += np.square(items[i]-ic[i])
                else:
                    sum += np.square(items[i])
            for i in ic.keys():
                if i not in  items:
                    sum += np.square(-ic[i])
                    
            odirec = np.sqrt(sum)       
            C[user].setdefault(uc,0)
            C[user][uc] =  odirec
            
    return C          
```

### 2.3 推荐

```python
def UCF_Recommend(user_id, user_item_pd, C, K=3,N=10):
    # user_id,需要推荐的用户  user_item_pd 用户商品二维矩阵 C，用户关系矩阵  K 选取相关用户的前K个商品，N推荐N个商品
    rank = {}
    action_item = user_item_pd[user_id]
    #根据user_id对应的列排序，得到排序高的前K行对应用户的数据
    top_C = C_pd.sort_values(by=user_id , ascending=False)
    top_C = top_C.iloc[0:K]
    #print(top_C)
    #遍历排名高的前K个用户
    for index,row in top_C.iterrows():
        # 因为top_C是用户相关系数矩阵所以index是相关用户的userid
    #     print(index)
        #得到用户user_id 与index 的分数
        uscore = row[user_id]
        #print("uscore"+str(uscore))
        #得到当前相似用户打分高的前K条记录
        ui = user_item_pd.sort_values(by=index , ascending=False).iloc[0:K]
        #user_item_pd 是用户-物品打分相关矩阵，所以index2是item_id
        for index2,row2 in ui.iterrows():
    #         print(index2)
    #         print(row2[index])
            #得到此用户index 对当前商品的打分
            uiscore = row2[index]
            
            if  action_item[index2] >0:
                #print('yes')
                continue
            #print('no')
            if index2 not in rank.keys():
                rank.setdefault(index2, 0)
            #根据当前用户index对此商品的评分以及用户相关系数的乘积，得到需推荐用户user_id对应的item得分
            rank[index2] += uscore * uiscore
     
    return rank
```

### 3.2计算物品-物品的相似举证

```python
#夹角余弦(Cosine)
def cos_sim(vector_a, vector_b):
    """
    计算两个向量之间的余弦相似度
    :param vector_a: 向量 a 
    :param vector_b: 向量 b
    :return: sim
    """
#     vector_a = np.mat(vector_a)
#     vector_b = np.mat(vector_b)
#     num = float(vector_a * vector_b.T)
#     denom = np.linalg.norm(vector_a) * np.linalg.norm(vector_b)
#     cos = num / denom
#     sim = 0.5 + 0.5 * cos
    
    sim=np.dot(vector_a,vector_b)/(np.linalg.norm(vector_a)*(np.linalg.norm(vector_b)))
    return sim
```

```python
#欧氏距离
def ou_norm(vector1, vector2):
    #op1=np.sqrt(np.sum(np.square(vector1-vector2)))
    op2=np.linalg.norm(vector1-vector2)
    return op2
```

```python
#曼哈顿距离(Manhattan Distance)
def manhattan(vector1, vector2):
    #op3=np.sum(np.abs(vector1-vector2))
    op4=np.linalg.norm(vector1-vector2,ord=1)
    return op4
```

```python
#切比雪夫距离(Chebyshev Distance)
def chebyshev(vector1, vector2):
    #op5=np.abs(vector1-vector2).max()
    op6=np.linalg.norm(vector1-vector2,ord=np.inf)
    return op6
```

#### 3.2.2 向量的方式

```python
def item_similarity2(item_data):
    C = dict()  #物品与物品的共现矩阵
    N = dict() #某物品被多少用户购买
    W = dict() #item的相似矩阵
    for user,items in item_data.items():
        #遍历每个用户对应的items
        for i in items.keys():
            #该物品被改用户购买的次数+1
            N.setdefault(i,0)
            N[i] +=1
            #新增一个共现矩阵
            C.setdefault(i,{})
            for j in items.keys():
                if i==j:
                    continue
                C[i].setdefault(j,0)
                C[i][j] +=  items[j]
                #print(items[j])
                
    for i,r_items in C.items():
        W.setdefault(i,{})
        for j,cij in r_items.items():
            #余弦相似度
            W[i][j] = cij/(math.sqrt(N[i]*N[j]))
    
    return W,C,N
```

```python
def ICF_recommend(user_predicate,user_item_dict,W,K=3,N=10):
    #user_predicate:需要推荐的用户，user:用户item举证 W：物品相似矩阵，K：每个item选取相似的前K个，N：最后结果返回N个
    rank = dict()
    #根据用户选取他有过动作的item合集
    count =0
    action_item = user_item_dict[user_predicate]
    for item,score in action_item.items():
        #遍历物品item最相似的前K个item。获得物品及分数
        count +=1
        if(count>2):
            break
        #print(item)
        #print(W[264960.0].items())
        #print(W[238805428.0].items())
        print(sorted(W[item].items(),key=lambda x:x[1],reverse = True))
        for j ,wj in sorted(W[item].items(),key=lambda x:x[1],reverse = True)[0:K]:
            #该物品为当前物品，不处理
            if  action_item[j] >0:
                continue
            rank.setdefault(j,0)
            rank[j]+= wj
    #return dict(sorted(rank.items(),key=lambda x:x[1],reverse = True)[0,N])
    if  rank:
        return dict(sorted(rank.items(),key=lambda x:x[1], reverse=True)[0:N])
    return rank
```

####  3.3.2矩阵的方式

```python
def ICF_recommend2(user_id,icf_user_item_pd,C_icf_pd,K=5,N=10):
    rank ={}
    #获取用户所有的item
    action_item = icf_user_item_pd[user_id]

    #获取用户相关度较高的item
    top_items = icf_user_item_pd.sort_values(by=user_id , ascending=False)
    top_item = top_items.iloc[0:K]
    #print(top_item)
    for index,row in top_item.iterrows():
        #index 用户对应的item_id, row 此item对应每个用户的打分
        #print(index)
        uscore = row[user_id]
        #print(uscore)
        #获取每个item相关系数最高的前N条item记录
        ii = C_icf_pd.sort_values(by=index , ascending=False).iloc[0:K]
       #user_item_pd 是用户-物品打分相关矩阵，所以index2是item_id,row为此item针对所有item的相关度
        for index2,row2 in ii.iterrows():
            #得到此用户index 对当前商品的打分
            iiscore = row2[index]
            if  action_item[index2] >0:
                continue
            if index2 not in rank.keys():
                rank.setdefault(index2, 0)
            #根据当前用户index对此商品的评分以及用户相关系数的乘积，得到需推荐用户user_id对应的item得分
            rank[index2] += uscore + iiscore
    
    return rank
```

```python
result2 = ICF_recommend2(10001082.0,icf_user_item_pd,C_icf_pd,N=1)
print(result2)
```
