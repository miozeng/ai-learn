#  K-近邻

https://github.com/apachecn/AiLearning
KNN是一个理论上比较成熟的方法，也是最简单的机器学习算法之一。
K-NN是一种基于实例的学习，
该方法的思路是：如果一个样本在特征空间中的k个最相似(即特征空间中最邻近)的样本中的大多数属于某一个类别，则该样本也属于这个类别。

概述：
KNN算法本身简单有效，他是一只lazy-learning算法
分类器不需要使用训练集进行训练，训练时间复杂度为0

KNN分类的计算复杂度和训练集中的文档数目成正比，也就是说如果训练集中文档的数目是n，那么KNN的分类时间复杂度没O(n)


优点：精度高、对异常值不敏感、无数据输入假定
缺点：计算复杂度高、空间复杂度高
适用数据范围：数值型和标称型

案列：
如果K=3，绿色圆点的最近的3个邻居是2个红色小三角形和1个蓝色小正方形，少数从属于多数，基于统计的方法，判定绿色的这个待分类点属于红色的三角形一类。
如果K=5，绿色圆点的最近的5个邻居是2个红色三角形和3个蓝色的正方形，还是少数从属于多数，基于统计的方法，判定绿色的这个待分类点属于蓝色的正方形一类。

计算过程：
计算已知类别数据与当前点的距离，
按距离排序
选取与当前距离最小的K个点
确定K个点所在类别的出现概率
返回前K个点出现频率最高的类别为当前点的类别

在KNN中，通过计算对象间距离来作为各个对象之间的非相似性指标，避免了对象之间的匹配问题，在这里距离一般使用欧氏距离或曼哈顿距离

### python实例 

电影可以按照题材分类，那么如何区分动作片和爱情片 呢？

动作片：打斗次数更多
爱情片：亲吻次数更多

```python
import numpy as np
import operator
```

#### 1.数据准备 

```python
 #四组二维特征。第一个是接吻的镜头数，第二个是打斗的镜头
group = np.array([[5,115],[7,106],[56,11],[66,9]])
#四组对应标签
labels = ('动作片','动作片','爱情片','爱情片')
#需要预测的数据
intX = [20,101]
```

#### 2.计算距离 

```python
#numpy中shape[0]返回数组的行数，shape[1]返回列数
dataSetSize = group.shape[0]

#3中方式计算欧几里得距离

diff = np.tile(intX, (dataSetSize, 1)) - group  # 按元素求差值
squaredDiff = diff ** 2  # 将差值平方
squaredDist = np.sum(squaredDiff, axis = 1)   # 按行累加
distance = squaredDist ** 0.5  # 将差值平方和求开方，即得距离
print(distance)
```

#### 3.按距离排序

```python

sortedDistIndices = np.argsort(distance)
print(sortedDistIndices)
```

#### 4.选取K个点 

```python
k = 3
```

#### 5.K个点出现类别的频率 

```python
classCount = {} # define a dictionary (can be append element)
for i in range(k):
    # # step 3: 选择k个最近邻
    voteLabel = labels[sortedDistIndices[i]]
    print(voteLabel)
    # # step 4: 计算k个最近邻中各类别出现的次数
    # when the key voteLabel is not in dictionary classCount, get()
    # will return 0
    classCount[voteLabel] = classCount.get(voteLabel, 0) + 1
print(classCount)
```

#### 6.返回前K个点出现频率最高的类别为当前点的类别 

```python
print(list(classCount.keys())[0] )
```
